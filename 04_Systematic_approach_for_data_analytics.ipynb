## Systematic Approach for Data Analysis

If you have access to pre-processed and clean data, you’ll typically take the following steps.

1. **Have your objective in mind** What do you want to find out? Are there metrics and criteria that are particularly important to you? bWhen would you be content with the results? 
2. **Initial data exploration and visualization**: Explore the data using visualization tools, draw initial assumptions about the relationships between variables, the importance of different features and the informative value of the data with respect to your questions and objectives. Humans are “visual beings”!
3. **Model design**: Choose models that you assume are able to answer the question. You’ll often compare the performance of several models and model configurations. 
    * Choose a model 
    * Select features and train the model
    * Evaluate the model’s performance 
    * Improve the model (e.g., via hyperparameter tuning)
4. **Making predictions**: Select one approach for making predictions / support decisions, or deploy the model in practice. 
5. **Interpretation and evaluation**: Challenge your results and do not trust them blindly! Do the conclusions seem reasonable? How reliable are they? Are there uncertainties that you have to accept and have to know about?

You may have to repeat one or several of these steps. Sometimes it may be necessary to start over, e.g. if your data ends up being insufficient for answering your question (in the desired quality).

## Preparing our data

'Manually' training and using a linear model requires a bit more effort than simply calling `seaborn`'s lmplot. So let's do this step by step!

First, we need to clean our data! Remember, that there are rows in our dataset where data is missing (`NaN`). Let's drop them. The following command does just that -- it drops entire rows with one or more `NaN`-values. The parameter `inplace` makes sure that our variable is edited in place, otherwise, the method would return a new DataFrame!

```python
penguins.head()
```

```python
penguins.isnull().sum()
```

```python
penguins.dropna?
```

```python
penguins.dropna(inplace=True)
penguins
```

```python
penguins.isnull().sum()
```

Our dataset is now shorter because any NaN-rows are gone.

Now, we'll repeat how to "select"/"extract" data from our DataFrame.

Type the name of a column in square brackets to select a column. This returns a `Series`. Remember: You can imagine a Series as an indexed one-dimensional array of values.

```python
penguins['body_mass_g']
```

If you select multiple columns (as a list of column names), you get a new `DataFrame`.

```python
penguins[['body_mass_g', 'flipper_length_mm']]
```

**Attention:** Some function expect a *DataFrame*, not a *Series*, even if there is just one column present. To achieve that, pass a list with just one value -- e.g. use two square brackets `[[]]`! 

General rule: `X` (the input data) is a DataFrame (multi-dimensional), `y` (the target value) is a Series!

This is exactly the case for the `fit`-function of our linear regression model!

## We train our first model

Now let's train our first model. We assume a linear relationship between the flipper length and the body mass. So we can use a Linear Regression model.

```python
# we 'initialize'/'construct' a new model
model = LinearRegression()
```

```python
# set variables X and y
X = penguins[['flipper_length_mm']]
y = penguins['body_mass_g']
```

```python
# then we fit/'train' the model using its fit-method
model.fit(X, y)
```

Let's take a closer look at this!

We pass two parameters to the `fit`-method of our Linear Regression Model. *Attention:* The first value must be a `DataFrame` (therefore, two square brackets!) and the second value must be a `Series` (therefore, one square bracket!). ((The reason is that the Linear Regression Model can have 1 or more inputs to create exactly 1 output value. Here we just have 1 input (flipper length) to guess the body mass!))

Internally, the model has been fitted using the least squares method. Let's look at two of it's properties.

```python
model.coef_
```

```python
model.intercept_
```

What do these values mean? To put it very simple: Our model guesses that you *"need to multiply the flipper length by 50 and substract 5872 to get an estimation of the body mass based on the flipper length"*.

Let's try it with the model's `predict`-method. Assume, that we have measured 200 mm flipper length and want to guess the weight. (Again: `predict` expects a multi-dimensional array because it can accept more than 1 input.)

```python
model.predict?
```

```python
model.predict([[200]])
```

This looks like a reasonable mass for a penguin with a flipper length of 200 mm. Check the pair plot above (flipper length vs body mass).

Let's look at the models' score:

```python
model.score?
```

```python
model.score(X, y)
```

## Manually plot the results

Let's plot the line our model predicted and check if it makes sense. (Again: Watch out for the double square brackets as `predicts` expects a DataFrame!) 

```python
# we start off with a scatter plot
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g")
# and then we plot a line on the same figure
sns.lineplot(x=X['flipper_length_mm'], y=model.predict(X))
```

### Bonus 1: Linear Regression with more than one dimension!
Let's create a model which accepts more than one input dimension.

```python
penguins
```

```python
# new: separate dataset for training and testing
from sklearn.model_selection import train_test_split

# prepare data as seen above 
penguins.dropna(inplace=True)

# new: multiple numerical inputs (more than one dimenson)
X = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']]
y = penguins['body_mass_g']
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
